```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
# r = requests.get("https://www.century21.com/for-sale-homes/Jacksonville-FL-373c")
c = r.content
```

```python
soup = BeautifulSoup(c, "html.parser")
```

```python
all = soup.find_all("div", {"class": "propertyRow"})
```

```python
all[0].find("h4", {"class":"propPrice"}).text.strip('\n').strip('$')
```

```python
l = []
for item in all:
    d={}

    d["Address"] = item.find_all("span",{"class":"propAddressCollapse"})[0].text.replace(".","")
    d["Locality"] =item.find_all("span",{"class":"propAddressCollapse"})[1].text
    d["Price"] = item.find("h4", {"class":"propPrice"}).text.strip('\n').strip('$')

    try:
        d["Beds"] = item.find("span", {"class":"infoBed"}).find("b").text
    except:
        d["Beds"] = None
    try:
        d["Area"] = item.find("span", {"class":"infoSqFt"}).find("b").text
    except:
        d["Area"] = None
    try:
        d["Full Baths"] = item.find("span", {"class":"infoValueFullBath"}).find("b").text
    except:
        d["Full Baths"] = None
    try:
        d["Half Baths"] = item.find("span", {"class":"infoValueHalfBath"}).find("b").text
    except:
        d["Half Baths"] = None
        
    for column_group in item.find_all("div", {"class","columnGroup"}):
        for feature_group, feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}), column_group.find_all("span",{"class":"featureName"})):
            if "Lot Size" in feature_group.text:
                d["Lot Size"] = feature_name.text.replace(",", "")
    
    l.append(d)
```

```python
data = pd.DataFrame(l)
```

```python
data.to_csv("data.csv", index=False)
```

```python
base_url = "http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/#t=0&s="
l = []

for page in range(0, 50, 10):
    r = requests.get(base_url + str(page) + ".html")
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    all = soup.find_all("div", {"class": "propertyRow"})
    for item in all:
        d={}

        d["Address"] = item.find_all("span",{"class":"propAddressCollapse"})[0].text.replace(".","")
        try:
            d["Locality"] = item.find_all("span",{"class":"propAddressCollapse"})[1].text
        except:
            d["Locality"] = None
        d["Price"] = item.find("h4", {"class":"propPrice"}).text.strip('\n').strip('$')

        try:
            d["Beds"] = item.find("span", {"class":"infoBed"}).find("b").text
        except:
            d["Beds"] = None
        try:
            d["Area"] = item.find("span", {"class":"infoSqFt"}).find("b").text
        except:
            d["Area"] = None
        try:
            d["Full Baths"] = item.find("span", {"class":"infoValueFullBath"}).find("b").text
        except:
            d["Full Baths"] = None
        try:
            d["Half Baths"] = item.find("span", {"class":"infoValueHalfBath"}).find("b").text
        except:
            d["Half Baths"] = None

        for column_group in item.find_all("div", {"class","columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}), column_group.find_all("span",{"class":"featureName"})):
                if "Lot Size" in feature_group.text:
                    d["Lot Size"] = feature_name.text.replace(",", "")

        l.append(d)
```

```python
data = pd.DataFrame(l)
```
